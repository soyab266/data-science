```python
rating_with_name= rating.merge(books ,on='ISBN')
```

```python
rating_with_name.head(4)
```

```python
num_rating_df=rating_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=rating_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df=num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df=popular_df[popular_df['num_ratings']>250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df=popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','num_ratings', 'avg_ratings','Image-URL-S']]
```

```python
popular_df
```

```python
x=rating_with_name.groupby('User-ID').count()['Book-Rating']> 200
educated= x[x].index
```

```python
filtered_rating=rating_with_name [rating_with_name['User-ID'].isin(educated)]
```

```python
filtered_rating.head(2)
```

```python
y=filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt=final_ratings.pivot_table( index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
similarity_score=cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
  index=np.where(pt.index==book_name)[0][0]
  similar_item=sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1], reverse=True)[1:6]
  #distance=similarity_score[index]
  for i in similar_item:
    print(pt.index[i[0]])

  
```

```python
recommend('1984')
```
