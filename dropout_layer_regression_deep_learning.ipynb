```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
pd.set_option('display.max_rows',None)
pd.set_option('display.max_columns',None)
import keras
#from sklearn.datasets import make_circles
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.layers import Dropout
```

```python
X_train=np.linspace(-1,1,20)
```

```python
X_train
```

```python
y_train=np.array([-0.6561 , -0.3099 , -0.59035, -0.50855, -0.285  ,
                    -0.2443 , -0.02445,  0.00135, -0.2006 ,  0.07475,
                    -0.1422 ,  0.06515,  0.15265,  0.3521 ,  0.28415,
                    0.5524 ,  0.23115,  0.20835, 0.4211,  0.60485])
```

```python
X_test=np.linspace(-1,1,20)
```

```python
X_test
```

```python
y_test=np.array([-0.69415, -0.451  , -0.43005, -0.4484 , -0.1475 ,
                   -0.5019 , -0.28055,  0.24595, -0.21425, -0.0286 ,
                   0.23415,  0.46575, 0.07955,  0.1973 ,  0.0719 ,
                   0.3639 ,  0.5536 ,  0.3365 , 0.50705,  0.33435])
```

```python
plt.scatter(X_train,y_train,c='b')
plt.scatter(X_test,y_test,c='g')
```

```python
model=Sequential()
model.add(Dense(256,activation='relu',input_dim=1))
model.add(Dense(256,activation='relu'))
model.add(Dense(1,activation='linear'))

adam=Adam(learning_rate=0.01)

model.compile(optimizer=adam,loss='mse', metrics=['mse'])
history=model.fit(X_train,y_train,epochs=500,verbose=True, validation_data=(X_test,y_test))
```

```python
y_pred=model.predict(X_test)
```

```python
plt.scatter(X_train,y_train)
plt.scatter(X_test,y_test)
plt.plot(X_test,y_pred)
```

```python
plt.plot(history.history['val_loss'])
plt.plot(history.history['loss'])
```

```python
model_1=Sequential()
```

```python
model_1.add(Dense(256,activation='relu',input_dim=1))
model_1.add(Dropout(0.4))
model_1.add(Dense(256,activation='relu'))
model_1.add(Dropout(0.4))
model_1.add(Dense(1,activation='linear'))
adam=Adam(learning_rate=0.01)
model_1.compile( optimizer=adam,loss='mse',metrics=['mse'])
history=model_1.fit(X_train,y_train,epochs=500,verbose=True,validation_data=(X_test,y_test))
```

```python
y_pred_1=model_1.predict(X_test)
```

```python
plt.scatter(X_train,y_train)
plt.scatter(X_test,y_test)
plt.plot(X_train,y_pred_1)
```
