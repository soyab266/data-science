```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
import requests
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.metrics import r2_score
from sklearn.linear_model import Lasso
```

```python
x,y=make_regression( n_samples=100,n_features=1,n_targets=1, n_informative=1, noise=25)
```

```python
x_train, x_test, y_train, y_test=train_test_split( x,y, test_size=0.2, random_state=20)
lr=LinearRegression()
lr.fit(x_train, y_train)
y_pred=lr.predict(x_test)
r2_score(y_pred, y_test)
```

```python
plt.scatter(x,y)
```

```python
alphas=[0,5,10,30]
plt.figure(figsize=(10,12))
for i in alphas:
  l=Lasso(alpha=i)
  l.fit(x_train, y_train)
  plt.scatter(x,y)
  plt.plot(x_test, l.predict(x_test),label='alpha={}'.format(i))
plt.legend()
plt.show()

```

lasso is known as l1 regularization , main difference  between ridge and lasso regularization is that  lasso creates sparcity(means zero value of coef), coef become zero for larger value of lamda(alpha).because of this phenomena feature selection is possible(some of the  value of columns become zero, which is not imp.)
