```python
pro= np.random.binomial(n=10,p=0.5,size=1000)
pro
```

IF WE INCREASE THE PROBABILITY OF GETTING HEAD THEN THE GRAPH WILL SKEW TO RIGHT SIDE. IE WE HAVE INCRESE THE PROBABILITY OF HEAD TO 0.8.

```python
n=10
p=0.8
size=1000
```

```python
pro=np.random.binomial(n,p,size)
```

```python
plt.hist(pro)
```

IF WE DECREASE THE PROBABILITY OF GETTING HEAD THEN THE GRAPH WILL SKEW TO LEFT SIDE. IE WE HAVE DECREASE THE PROBABILITY OF HEAD TO 0.2.

```python
n=10
p=0.2
size=1000
```

```python
pro=np.random.binomial(n,p,size)
pro
```

```python
plt.hist(pro)
plt.show()
```
