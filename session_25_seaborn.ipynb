### 1. Relational Plot

- to see the statistical relation between 2 or more variables.
- Bivariate Analysis

Plots under this section

- scatterplot
- lineplot

```python
# scatter plot
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(data=tips, x='tip' , y='total_bill')
```

```python
gap=px.data.gapminder()
gap.head(2)
```

```python
temp=gap[gap['country']=='India']
temp
```

```python
sns.lineplot(data=temp, x='year', y='lifeExp')
```

```python
sns.relplot(data=temp, x='year', y='lifeExp', kind='line', hue=)
```

```python
temp=gap[gap['country'].isin(['India', 'Brazil', 'Germany'])]
temp
```

```python
sns.relplot(data=temp, x='year', y='lifeExp', hue='country', kind='line',style='continent')
```

```python
sns.lineplot(data=temp, x='year', y='lifeExp', hue='country')
```

```python
# scatter using relplot -> size and hue
```

```python
sns.relplot(data=tips, x='total_bill', y='tip', col='sex', row='smoker', kind='scatter')
```

```python
gap.head()
```

```python
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=4)
```

### 2. Distribution Plots

- used for univariate analysis
- used to find out the distribution
- Range of the observation
- Central Tendency
- is the data bimodal?
- Are there outliers?

Plots under distribution plot

- histplot
- kdeplot
- rugplot

```python
# plotting univariate histogramsns
```

```python
tips
```

```python
sns.histplot(data=tips, x='total_bill')
```

```python
sns.displot(data=tips, x='day', kind='hist',hue='sex')
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
```

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

```python
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

```python
temp=gap.pivot(index='country', columns='year', values='lifeExp')
```

```python
sns.heatmap(data=temp, cmap='winter',)
```

```python
temp=gap[gap['continent']=='Europe'].pivot(index='country', columns='year', values='lifeExp')
plt.figure(figsize=(10,8))
sns.heatmap(data=temp, annot=True, cmap='winter')
```
